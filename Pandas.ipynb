## Creating a DataFrame

A DataFrame is a two-dimensional labeled data structure with columns of potentially different types.  
You can create a DataFrame from a dictionary, a list of lists, or other data structures.

Below is the first example from the article.

```python
# Example adapted from the PandasIntroduction.html article (UNIGIS Webbook)

import pandas as pd

data = {
    "Name": ["John", "Anna", "Peter"],
    "Age": [28, 24, 35],
    "Country": ["USA", "Germany", "Canada"]
}

df = pd.DataFrame(data)
df    # Ausgabe statt print(df)
```
